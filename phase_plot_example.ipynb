# Phase plot example

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
def lotka_volterra(R, F, p):
    alpha, beta, gamma, delta = p
    dRdt = alpha * R - beta * R * F
    dFdt = delta * beta * R * F - gamma * F
    return (dRdt, dFdt)
```

```python
rabbit_range = np.arange(0.1, 4.5, 0.2)
fox_range = np.arange(0.1, 4.5, 0.2)

# make grid
R, F = np.meshgrid(rabbit_range, fox_range)
alpha, beta, gamma, delta = 0.12, 0.07, 0.09, 0.8
p = (alpha, beta, gamma, delta)

dR, dF = lotka_volterra(R, F, p)
```

```python
vector_lengths = np.sqrt(dR**2 + dF**2)
```

```python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111)
qplot = ax.quiver(R, F, dR/vector_lengths, dF/vector_lengths, vector_lengths, cmap='jet')
ax.set_xlabel('Billions of rabbits', fontsize=20)
ax.set_ylabel('Billions of foxes', fontsize=20)
ax.set_title('Phase space', fontsize=25)
ax.tick_params(labelsize=15)
cbar = plt.colorbar(qplot, ax=ax)
cbar.ax.tick_params(labelsize=15)
plt.show()
```
